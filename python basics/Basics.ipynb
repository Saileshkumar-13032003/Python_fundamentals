```python
# Program to reverse a number using while loop
num = 12345
reversed_num = 0

while num != 0:
    digit = num % 10          # 1. Get the last digit
    reversed_num = reversed_num * 10 + digit  # 2. Append digit to the end
    num //= 10                # 3. Remove the last digit

print("Reversed Number: " + str(reversed_num))
```

```python
# Simple Triangle Classifier with Validity Check

s1 = float(input("Enter first side: "))
s2 = float(input("Enter second side: "))
s3 = float(input("Enter third side: "))

# Check if the sides can actually form a triangle
if (s1 + s2 > s3) and (s1 + s3 > s2) and (s2 + s3 > s1):
    
    if s1 == s2 == s3:
        print("Equilateral Triangle")
        
    elif s1 == s2 or s2 == s3 or s1 == s3:
        print("Isosceles Triangle")
        
    else:
        print("Scalene Triangle")

# If the first 'if' condition fails, it is not a valid triangle
else:
    print("Invalid Triangle")
```

```python
#scope of variable
1.local
2.global
3.non-local 
```

```python
def local():
    x=10
    print(x)
local()
```

```python
z=19

def var():
    print(z)
var()
print(z)
```

```python
z=19

def var():
    global z
    z=10
    print(z)
var()
print(z)
```

```python
def outer():
    w="outer"
    def inner():
        nonlocal w
        w="changed in inner"
    inner()
    print(w)
outer()
```

```python
try:
    z=int(input("Enter the number"))
    y=z/0
    print(y)
except Exception as e:
    print(e)
finally:
    print("Always executed")
```
